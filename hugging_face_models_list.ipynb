```python
from huggingface_hub import HfApi
import requests
```

```python
# Define function to get updated model list
def update_model_list():
    
    #Get list of all models available under Helsinki-NLP org
    api = HfApi()
    model_list = api.list_models(author="Helsinki-NLP")
    
    #Filter out models for "big"
    available_models = [model.modelID for model in model_list if "opus-mt" in model.modelID]
    
    return available_models

available_model_names = update_model_list
```

```python


def fetch_helsinki_nlp_models(prefix="Helsinki-NLP/opus-mt-"):
    models = []
    # Hugging Face Hub API endpoint for models
    url = "https://huggingface.co/api/models"
    # Parameters for the request, specifying the organization
    params = {
        "search":"opus-mt-tc-big",
        "author": "Helsinki-NLP",
        "full": False,
        "config":False # You can set this to True if you want more information about each model
    }
    
    # Make a GET request to fetch the models
    response = requests.get(url, params=params)
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the response JSON to get the list of models
        data = response.json()
        for model in data:
            if model['id'].startswith(prefix):
                 models.append(model['id'])
        print(models)

    else:
        print("Failed to retrieve models:", response.status_code)
    return models

# Call the function to print the Helsinki-NLP models
helsinki_models = fetch_helsinki_nlp_models()
print(helsinki_models)
```

```python
params_big = {
        "search":"opus-mt-tc-big",
        "author": "Helsinki-NLP",
        "full": False,
        "config":False
}

params_mt = {
        "search":"opus-mt-",
        "author": "Helsinki-NLP",
        "full": False,
        "config":False
}

def fetch_helsinki_nlp_models():
    models = []
    # Hugging Face Hub API endpoint for models
    url = "https://huggingface.co/api/models"

    # Make a GET request to fetch the models
    response = requests.get(url, params=params_big)
    response2 = requests.get(url, params=params_mt)
    # Check if the request was successful
    if response.status_code == 200:
        # Parse the response JSON to get the list of models
        data = response.json()
        for model in data:
            models.append(model['id'])
        # print(models)
        
    else:
        print("Failed to retrieve models:", response.status_code)
    
    if response2.status_code == 200:
    # Parse the response JSON to get the list of models
        data2 = response2.json()
        for model in data2:
            models.append(model['id'])
        # print(models)
    
    else:
        print("Failed to retrieve models:", response2.status_code)
    return models


# Call the function to print the Helsinki-NLP models
helsinki_models = fetch_helsinki_nlp_models()
print(helsinki_models)
```

```python
#Define function to save model names to txt file
def save_model_names_to_file(model_names, filename="helsinki_models.txt"):
    with open(filename, 'w') as file:
        for model in model_names:
            file.write(model + '\n')
            
save_model_names_to_file(helsinki_models)
```
